# **Import Library**

```python
import pandas as pd
import numpy as np
```

# **Separating User and Tour Guide Data**

## Tour Guide Data

```python
df = pd.read_csv('data/TRAVEL-PREFERENCES-AND-PERSONALITY-MATCHING-SURVEY.csv')
df
```

### Selecting Tour Guide Data

```python
# Looking at the index of the columns
columns = df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Tour Guide Dataset
tg_df = df[~((df.iloc[:, 2] == 'No') | (df.iloc[:, 6] == 'Other'))]
tg_df
```

```python
# Deleting unused columns with NaN value
tg_df = tg_df.copy()
tg_df.dropna(axis=1, how='all', inplace=True)
```

```python
# Looking at the index of tg_df columns
columns = tg_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Drop unused columns for applying distance metric
tg_df = tg_df.copy()
columns_to_drop = [0, 1, 2]

tg_df.drop(tg_df.columns[columns_to_drop], axis = 1, inplace = True)
tg_df
```

```python
# Rename tg_df columns
tg_df = tg_df.rename(columns={'Please indicate your top three areas of expertise from the following options': 'Expertise'})
tg_df = tg_df.rename(columns={tg_df.columns[5:10][i]: f'EXT{i+1}' for i in range(5)})
tg_df = tg_df.rename(columns={tg_df.columns[10:15][i]: f'EST{i+1}' for i in range(5)})
tg_df = tg_df.rename(columns={tg_df.columns[15:20][i]: f'AGR{i+1}' for i in range(5)})
tg_df = tg_df.rename(columns={tg_df.columns[20:25][i]: f'CSN{i+1}' for i in range(5)})
tg_df = tg_df.rename(columns={tg_df.columns[25:30][i]: f'OPN{i+1}' for i in range(5)})
tg_df
```

### One Hot Encoding for Tour Guide Data

```python
# Gender
# 'Male' = 0 dan 'Female' = 1
tg_df['Gender'] = tg_df['Gender'].map({'Male': 0, 'Female': 1})
```

```python
# Age
tg_df['17-25'] = tg_df['Age'].apply(lambda x: 1 if '17-25' in x else 0)
tg_df['26-34'] = tg_df['Age'].apply(lambda x: 1 if '26-34' in x else 0)
tg_df['35-43'] = tg_df['Age'].apply(lambda x: 1 if '35-43' in x else 0)
tg_df['44-52'] = tg_df['Age'].apply(lambda x: 1 if '44-52' in x else 0)
tg_df['52+'] = tg_df['Age'].apply(lambda x: 1 if '52+' in x else 0)

# Deleting 'Age' column 
tg_df.drop('Age', axis=1, inplace=True)
```

```python
# Domicile
# 'Bandung' = 0 dan 'Makassar' = 1
tg_df['Domicile'] = tg_df['Domicile'].map({'Bandung': 0, 'Makassar': 1})
```

```python
# Expertise
tg_df['Historical tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Historical tours' in x else 0)
tg_df['Adventure tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Adventure tours' in x else 0)
tg_df['Nature and wildlife tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Nature and wildlife tours' in x else 0)
tg_df['Culinary tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Culinary tours' in x else 0)
tg_df['Wellness and retreat tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Wellness and retreat tours' in x else 0)
tg_df['Architectural tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Architectural tours' in x else 0)
tg_df['Educational tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Educational tours' in x else 0)
tg_df['Shopping tours'] = tg_df['Expertise'].apply(lambda x: 1 if 'Shopping tour' in x else 0)

# Deleting 'Expertise' column 
tg_df.drop('Expertise', axis=1, inplace=True)

tg_df
```

```python
# Looking at the index of tg_df columns
columns = tg_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Check Value before reorder column
tg_df['17-25'].value_counts()
```

```python
# Reorder Column Position
cols = tg_df.columns.tolist()
new_cols = cols[:3] + cols[28:] + cols[3:28]
tg_df = tg_df[new_cols]

tg_df
```

```python
# Check Column Order
tg_df.columns
```

```python
# Check The Value isn't change after reorder column
tg_df['17-25'].value_counts()
```

### Save Dataframe into CSV as Tour Guide Datasets

```python
tg_df
```

```python
filepath = 'data/tour-guide-data.csv'
tg_df.to_csv(filepath, index=False)
```

## Tourist Data

### Selecting Tourist Data

```python
# Tour Guide Dataset
tourist_df = df[((df.iloc[:, 2] == 'No') | (df.iloc[:, 6] == 'Other'))]
tourist_df
```

```python
# Deleting unused columns with NaN value
tourist_df = tourist_df.copy()
tourist_df.dropna(axis=1, how='all', inplace=True)
```

```python
# Looking at the index of tourist_df columns
columns = tourist_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Drop unused columns for applying distance metric
tourist_df = tourist_df.copy()
columns_to_drop = [0, 1, 2]

tourist_df.drop(tourist_df.columns[columns_to_drop], axis = 1, inplace = True)
tourist_df
```

```python
# Rename tourist_df columns
tourist_df = tourist_df.rename(columns={'Please select your top three preferred tour themes from the following options': 'Tour preference.1'})
tourist_df = tourist_df.rename(columns={tourist_df.columns[9:14][i]: f'EXT{i+1}' for i in range(5)})
tourist_df = tourist_df.rename(columns={tourist_df.columns[14:19][i]: f'EST{i+1}' for i in range(5)})
tourist_df = tourist_df.rename(columns={tourist_df.columns[19:24][i]: f'AGR{i+1}' for i in range(5)})
tourist_df = tourist_df.rename(columns={tourist_df.columns[24:29][i]: f'CSN{i+1}' for i in range(5)})
tourist_df = tourist_df.rename(columns={tourist_df.columns[29:34][i]: f'OPN{i+1}' for i in range(5)})
tourist_df
```

```python
# Looking at the index of tourist_df columns
columns = tourist_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Combine the columns into a new column
tourist_df['Names'] = tourist_df['Name'].fillna(tourist_df['Name.1'])
tourist_df['Genders'] = tourist_df['Gender'].fillna(tourist_df['Gender.1'])
tourist_df['Ages'] = tourist_df['Age'].fillna(tourist_df['Age.1'])
tourist_df['Destinations'] = tourist_df['Domicile'].fillna(tourist_df['Domicile.1'])
tourist_df['Tour Preferences'] = tourist_df['Tour preference'].fillna(tourist_df['Tour preference.1'])

# Drop the original columns
columns_to_drop = [0, 1, 2, 3, 4,
                   5, 6, 7, 8, 34]

tourist_df.drop(tourist_df.columns[columns_to_drop], axis = 1, inplace = True)
tourist_df
```

```python
# Looking at tourist_df columns
tourist_df.columns
```

```python
# Check that we not missing any value after combine columns
print('Is there any missing value? ', tourist_df.isnull().values.any())
print('How many missing values? ', tourist_df.isnull().values.sum())
```

### One Hot Encoding for Tourist Data

```python
# Genders
# 'Male' = 0 dan 'Female' = 1
tourist_df['Genders'] = tourist_df['Genders'].map({'Male': 0, 'Female': 1})
```

```python
# Ages
tourist_df['17-25'] = tourist_df['Ages'].apply(lambda x: 1 if '17-25' in x else 0)
tourist_df['26-34'] = tourist_df['Ages'].apply(lambda x: 1 if '26-34' in x else 0)
tourist_df['35-43'] = tourist_df['Ages'].apply(lambda x: 1 if '35-43' in x else 0)
tourist_df['44-52'] = tourist_df['Ages'].apply(lambda x: 1 if '44-52' in x else 0)
tourist_df['52+'] = tourist_df['Ages'].apply(lambda x: 1 if '52+' in x else 0)

# Deleting 'Ages' column 
tourist_df.drop('Ages', axis=1, inplace=True)
```

```python
# Map 'Other' to a random value between 'Bandung' and 'Makassar'
random_destinations = np.random.choice(['Bandung', 'Makassar'], size=tourist_df['Destinations'].eq('Other').sum())
tourist_df.loc[tourist_df['Destinations'] == 'Other', 'Destinations'] = random_destinations
```

```python
# Check map value result
tourist_df['Destinations'].value_counts()
```

```python
# Destinations
# 'Bandung' = 0, 'Makassar' = 1
tourist_df['Destinations'] = tourist_df['Destinations'].map({'Bandung': 0, 'Makassar': 1})
```

```python
# Tour Preferences
tourist_df['Historical tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Historical tours' in x else 0)
tourist_df['Adventure tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Adventure tours' in x else 0)
tourist_df['Nature and wildlife tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Nature and wildlife tours' in x else 0)
tourist_df['Culinary tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Culinary tours' in x else 0)
tourist_df['Wellness and retreat tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Wellness and retreat tours' in x else 0)
tourist_df['Architectural tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Architectural tours' in x else 0)
tourist_df['Educational tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Educational tours' in x else 0)
tourist_df['Shopping tours'] = tourist_df['Tour Preferences'].apply(lambda x: 1 if 'Shopping tour' in x else 0)

# Deleting 'Tour Preferences' column 
tourist_df.drop('Tour Preferences', axis=1, inplace=True)

tourist_df
```

```python
# Looking at the index of tourist_df columns
columns = tourist_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
# Reorder the columns
cols = tourist_df.columns.tolist()
new_cols = cols[25:] + cols[:25]
tourist_df = tourist_df[new_cols]
```

```python
# Looking at tourist_df columns
tourist_df.columns
```

```python
# check if the value is not change
tourist_df['Destinations'].value_counts()
```
