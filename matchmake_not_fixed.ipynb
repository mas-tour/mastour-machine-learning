# **Make Matchmaking Model**

## Reduce Dimension from Data

```python
match_array = match_df.values
match_array
```

## Test Load Model

```python
# Load the PCA model
loaded_model = tf.saved_model.load('pca_model')
```

```python
new_data_tf = tf.convert_to_tensor(match_array, dtype=tf.float32)

# Predict the PCA value using the loaded model
pca_value = loaded_model.apply_pca(new_data_tf)

# Print the PCA value
print(pca_value)
```

```python
pca_value_arr = pca_value.numpy()
pca_value_arr
```

```python
pca_result_df = pd.DataFrame(pca_value_arr)
pca_result_df
```

```python
# Apply K-means clustering
num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, n_init='auto')
kmeans.fit(pca_value_arr)

# Get the cluster labels for each user
user_clusters = kmeans.labels_

# Plot the clustering result in the reduced feature space
plt.scatter(pca_value_arr[:, 0], pca_value_arr[:, 1], c=user_clusters, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clustering Result')
plt.show()
```

```python
unique_values, counts = np.unique(user_clusters, return_counts=True)

# Print the unique values and their respective counts
for value, count in zip(unique_values, counts):
    print(f'{value}: {count}')
```

```python
# Looking at the index of the columns
columns = user_df.columns
for idx, column in enumerate(columns):
    print(f"Column index: {idx}, Column name: {column}")
```

```python
def matchmaking(user_index, num_recommendations=6):
    # Get the cluster label for the specified user
    user_cluster = user_clusters[user_index]

    # Get the indices of the users within the same cluster as the specified user
    cluster_indices = [i for i, cluster_label in enumerate(user_clusters) if cluster_label == user_cluster]

    # Get the distances within the cluster of the specified user
    cluster_distances = euclidean_distances(pca_value_arr[cluster_indices])

    # Get the distances for the specified user
    user_distances = cluster_distances[cluster_indices.index(user_index)]

    # Normalize the distances within the cluster to get the scores within the range of 0-100
    normalized_distances = 1 - user_distances / np.max(user_distances)
    scores = normalized_distances * 100

    # Get the indices and scores of the closest matches within the cluster
    matches = [(index, score) for index, score in enumerate(scores) if index != user_index and score > 25]

    # Sort the matches based on the highest score
    matches.sort(key=lambda x: x[1], reverse=True)

    print("User Index:", user_index)
    print("Matched Peers:")
    match_idx = [user_index]
    for match_index, score in matches:
        if user_df.iloc[cluster_indices[match_index], 21] == 'guide':
            if user_df.iloc[cluster_indices[match_index], 2] == user_df.iloc[user_index, 2]:
                match_idx.append(cluster_indices[match_index])
                print("Index:", cluster_indices[match_index], "| Score:", f"{score:.2f}%")

                if len(match_idx) == num_recommendations:
                    break

    # Check if the number of recommended guides is less than the desired number
    if len(match_idx) < num_recommendations:
        remaining_recommendations = num_recommendations - len(match_idx)
        
        # Get the indices of users in other clusters with the same destination
        other_cluster_indices = [i for i, cluster_label in enumerate(user_clusters) if cluster_label != user_cluster and score > 25]
        other_cluster_indices = [idx for idx in other_cluster_indices if user_df.iloc[idx, 2] == user_df.iloc[user_index, 2]]
        
        # Get the distances within the other clusters
        other_cluster_distances = euclidean_distances(pca_value_arr[other_cluster_indices])
        
        # Get the distances for the specified user
        user_distances_other = other_cluster_distances[:, cluster_indices.index(user_index)]
        
        # Normalize the distances within the other clusters to get the scores within the range of 0-100
        normalized_distances_other = 1 - user_distances_other / np.max(user_distances_other)
        scores_other = normalized_distances_other * 100
        
        # Get the indices and scores of the closest matches in other clusters
        other_matches = [(index, score) for index, score in enumerate(scores_other)]
        
        # Sort the matches based on the highest score
        other_matches.sort(key=lambda x: x[1], reverse=True)
        
        # Append the indices of guides from other clusters to the match_idx list
        for match_index, score in other_matches:
            if user_df.iloc[other_cluster_indices[match_index], 21] == 'guide':
                match_idx.append(other_cluster_indices[match_index])
                print("Index:", other_cluster_indices[match_index], "| Score:", f"{score:.2f}%")
                
                # Check if the desired number of recommendations is reached
                if len(match_idx) == num_recommendations:
                    break
    
    # Check if Destination and roles are correct 
    print()         
    print(user_df.iloc[match_idx, [0, 1, 2, 21]])
```
